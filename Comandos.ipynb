```python
#Calcular os Indicadores (Faturamnto e Quantidade de Produtos)
import pandas as pd
# caso tenha mais de uma aba no Excel, use o comando sheet_name=1 ou 2 ou 3...
tabela = pd.read_excel(r'C:\Users\tarci\OneDrive\Documentos\Cursos Curriculares\Automatização de Sistemas e Ciência de Dados\Aula 1 - Automação de Processos\Vendas - Dez.xlsx')
display(tabela)
faturamento = tabela['Valor Final'].sum()
qtde_produtos = tabela['Quantidade'].sum()
```
